```python
import matplotlib.pyplot as plt
```

```python
import csv
```

```python
x = []
y = []
with open("./data/eq2013.csv") as f:
    reader = list(csv.reader(f))
    for i in range(1,len(reader)):
        x.append(float(reader[i][2]))
        y.append(float(reader[i][3]))
```

```python
wordmap = plt.imread("./data/world.png")
```

## gridsize是XY方向网格的数量，
### 如果是一个参数，则表示X方向的数目，Y方向的数量自动计算
### 如果是两个参数，则表示手动设置xy方向各有多少个数目

## bins参数表示聚合的算法，这里用Log，表示log10(i+1)计数
## 默认就是直接计数

```python
plt.figure(figsize=(14, 10))
plt.imshow(wordmap, zorder=0, extent=[-179.99,179.999,-89.999,89.9999])
plt.hexbin(x,y,gridsize=(36,9),alpha=0.7,bins='log', cmap='inferno')
plt.show()
```

```python
plt.figure(figsize=(14, 10))
plt.imshow(wordmap, zorder=0, extent=[-179.99,179.999,-89.999,89.9999])
plt.hexbin(x,y,gridsize=(60,30),alpha=0.7,bins='log', cmap='inferno')
plt.show()
```

# hexbin聚合1000万个点绘制测试

```python
import random,datetime
```

```python
psize = 10000000
```

```python
x = [random.uniform(-180,180) for i in range(psize)]
y = [random.uniform(-90,90) for i in range(psize)]
```

```python
s = datetime.datetime.now()
plt.figure(figsize=(14, 10))
plt.hexbin(x,y,gridsize=(60,30),alpha=0.7,bins='log', cmap=plt.cm.jet)
plt.show()
print(datetime.datetime.now() -s)
```
