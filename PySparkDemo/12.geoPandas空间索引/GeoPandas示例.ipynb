```python
import random
```

```python
import geopandas as gpd
```

```python
from shapely.geometry import Point
```

## 构建一万个点

```python
parr = [Point(random.uniform(50,135),
           random.uniform(18,53)) for i in range(10000)]
```

```python
parr[:10]
```

## 利用GeoPandas直接读取shp文件，获得一个DataFrame

```python
cnpd = gpd.GeoDataFrame.from_file("./data/cn.shp")
```

```python
cnpd
```

```python
cnpd[cnpd["geometry"].intersects(parr[0])]["FIRST_NAME"].to_string(index=False)
```

```python
import datetime
```

## 测试一下，用点去查询这些面

```python
s = datetime.datetime.now()
for p in parr:
    cnpd["geometry"].intersects(p)
print(datetime.datetime.now()-s)
```

## 构建一个空间索引，再来试试

```python
spatial_index = cnpd.sindex
```

```python
ls = list(spatial_index.intersection(parr[0].bounds))
```

```python
pm = cnpd.iloc[ls]
```

```python
pm[pm.intersects(p)]
```

## 利用空间索引进行1000次查询

## 发现一个问题，如果是点对面的查询，空间索引的效率很不明显，因为构建索引也挺花时间的

```python
s = datetime.datetime.now()
for p in parr:
    ls = list(spatial_index.intersection(p.bounds))
    pm = cnpd.iloc[ls]
    pm.intersects(p)
print(datetime.datetime.now()-s)
```

## 传统迭代模式

## 效率的进步还是很明显的

```python
s = datetime.datetime.now()
for p in parr:
    for i in range(len(cnpd)):
        p.intersects(cnpd["geometry"][i])
print(datetime.datetime.now()-s)
```

# 当然，如果丢弃掉所有的属性数据，就走空间，
# 传统模式也不慢，主要是pandas的迭代本来就很强大

# 但是这种情况，就没有任何分析的意义了

```python
s = datetime.datetime.now()
cngeo = cnpd["geometry"]
for p in parr:
    for geo in cngeo:
        p.intersects(geo)
print(datetime.datetime.now()-s)
```
