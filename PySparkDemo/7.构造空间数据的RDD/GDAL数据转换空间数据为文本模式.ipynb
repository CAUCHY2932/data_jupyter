```python
import ogr
```

```python
shpfile = "./data/dltb.shp"
outjson = "./data/dltb.json"
outtsv = "./data/dltb.tsv"
```

# 转换为GeoJSON

## 在GDAL里面可以直接加载GeoJSON的驱动，进行转换

```python
shp = ogr.GetDriverByName("ESRI Shapefile").Open(shpfile)
ogr.GetDriverByName("GeoJSON").CopyDataSource(shp, outjson)
shp = None
```

## 读取输出的数据看看

```python
import json
```

```python
dltb = json.load(open(outjson,encoding="utf-8"))
```

# WKT模式

## GDAL可以直接将geometry解析成wkt字符串

```python
shp = ogr.GetDriverByName("ESRI Shapefile").Open(shpfile)
```

### GDAL读取数据，并且获取字段描述信息

```python
lyr = shp.GetLayerByIndex(0)
lyd = lyr.GetLayerDefn()
fields = [lyd.GetFieldDefn(i).GetName() for i in range(lyd.GetFieldCount())]
```

```python
fields
```

### 写入表头（字段名）

```python
head = "FID\tSHAPE\t"+"\t".join(fields)
print(head)
```

```python
w = open(outtsv,"w")
w.write(head+"\n")
```

### 写入数据

```python
for i in range(lyr.GetFeatureCount()):
    feat = lyr.GetFeature(i)
    feat.GetFID
    wkt = feat.GetGeometryRef().ExportToWkt()
    line = "{0}\t{1}".format(feat.GetFID(),wkt)
    for f in fields:
        line ="{0}\t{1}".format(line,feat.GetField(f))
    w.write(line+"\n")
```

```python
w.close()
```
