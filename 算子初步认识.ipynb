```python
import pyspark
```

```python
sc = pyspark.SparkContext()
```

```python
import random
```

```python
import datetime
```

```python
x = [chr(random.randint(65, 91)) for i in range(10000)]
```

```python
x
```

```python
s = datetime.datetime.now()
```

```python
rdd = sc.parallelize(x)
```

```python
rdd
```

```python
datetime.datetime.now()-s
```

```python
rdd.count()
```

```python
rdd2 = rdd.map(lambda x: (x, 1))
print(datetime.datetime.now() - s)
print(rdd2)
```

```python
s = datetime.datetime.now()
rdd3 = rdd2.reduceByKey(lambda a, b: a+b)
print(datetime.datetime.now() - s)
print(rdd3)
```

```python
s = datetime.datetime.now()
res = rdd3.collect()
print(datetime.datetime.now() - s)
print(res)
```
